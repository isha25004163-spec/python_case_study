```python
file=input("")
while(1):
  print("1. View Tasks")
  print("2. Update Task")
  print("3. Add Task")
  print("4. Delete Task")
  print("5. Search Task")
  print("6. Exit")
  try:
    ch=int(input())
  except:
    print("Invalid Choice")
    continue
  match ch:
    case 1:
      try:
        with open(file,'r') as f:
          print(f.read())
      except FileNotFoundError:
        print("File not found")
        continue

    case 2:
      val=input("Which task is completed: ")
      with open(file,'r') as f:
        content=f.readlines()
      except FileNotFoundError:
        print("File not found")
        continue

      found=False
      content=[i.strip().rsplit(' ',1) for i in content]
      for i in range(len(content)):
        if content[i][0].lower()==val.lower():
          content[i][1]='Done'
          found=True

      if not found:
        print("Not found")
      content=[i[0]+' '+i[1]+'\n' for i in content]
      with open(file,'w') as f:
        f.writelines(content)

    case 3:
      val=input("Enter the Task: ")
      cond=input("Enter the state of completion (Done/Not Done)")
      with open(file,'a') as f:
        f.write(val+' '+cond+'\n')
      except FileNotFoundError:
        print("File not found")
        continue

    case 4:
      val=input("Enter the task: ")
      with open(file,'r') as f:
        content=f.readlines()
      except FileNotFoundError:
        print("File not found")
        continue

      content=[i.strip().rsplit(' ',1) for i in content]
      content= [i for i in content if i[0].lower()!=val.lower()]
      content=[i[0]+' '+i[1]+'\n' for i in content]
      with open(file,'w') as f:
        f.writelines(content)

    case 5:
      val=input("Enter the Task: ")
      with open(file,'r') as f:
        content=f.readlines()
      except FileNotFoundError:
        print("File not found")
        continue

      content=[i.strip().rsplit(' ',1) for i in content]
      for i in range(len(content)):
        if content[i][0].lower()==val.lower():
          print(content[i][0],'-',content[i][1])
          break
      else :
        print("Task isn't found")
    case 6:
      break
```
