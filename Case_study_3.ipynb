```python
class Salary():
  def __init__(self,ID,name,basic_sal):
    self._ID=ID
    self.name=name
    self.basic_sal=basic_sal
    self._gross=0
    self._net=0
    self._tax=0
  def calculate_hra(self):
    return self.basic_sal*0.4
  def calculate_da(self):
    return self.basic_sal*0.6
  def gross_sal(self):
    self._gross=self.calculate_hra()+self.calculate_da()+self.basic_sal
    return self._gross
  def net_sal(self,tax):
    self._net=self.gross_sal()-(self._gross*tax)
    self._tax=tax
    return self._net
  def show_sal(self):
    print(f"ID: {self._ID}")
    print(f"Name: {self.name}")
    print(f"Basic Salary: {self.basic_sal}")
    print(f"HRA: {self.calculate_hra()}")
    print(f"DA: {self.calculate_da()}")
    print(f"Gross Salary: {self._gross}")
    print(f"Tax: {self._tax}")
    print(f"Net Salary: {self._net}")



```
